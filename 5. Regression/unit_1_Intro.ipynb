```python
# เตรียมเครื่องมือ 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# data preparation 
# create data ==> create ==> Dictionnay ==> Create Dataframe
x_data = [1, 2, 3, 4, 5, 6]   # ข้อมูล x แบบ List 
y_data = [1, 1.3, 2.2, 2.5, 2.8, 3.6] # ข้อมูล y  List 
x = np.array(x_data) # แปลงข้อมูล x เป็น arrray
y = np.array(y_data) # แปลงข้อมูล y เป็น array 

# dictionary 
data_dict = { "x": x_data, "y": y_data} # create dicxtionary 

# DataFrame
df = pd.DataFrame(data_dict) # create data frame 
```

```python
# สำรวจความสำพันธ์ โดนใช้กราฟ
plt.scatter(x,y)   # ใช้ข้อมูลจาก Numby array 
# plt.title(" X VS Y ") 
# plt.xlabel("X") 
# plt.ylabel("Y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()    
```

```python
x.shape, y.shape  # ตรวจสอบรูปร่าง ( ตรวจสอบ มิติ )  
```

```python
# ข้อมูลที่จะใช้ในการสร้าง ต้องอยู่ในรูปของ matrix / 2D row and column 
x = x.reshape(-1,1)
y = y.reshape(-1,1)
x.shape,y.shape
```

```python
x.shape, y.shape  # ตรวจสอบรูปร้าง ( ตรวจสอบ มิติ ) 
```

```python
# train model ฝึก 
model.fit(x,y)
```

```python
a = model.intercept_.round(4)  # จุดตัดบนแกน y   (y = a +bx) 
a
```

```python
b = model.coef_.round(2)    # ส.ป.ส. หน้าตัวแปร x  (y = a +bx) 
b
```
